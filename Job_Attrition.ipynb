```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

df = pd.read_csv("gdg_train.csv")
le = LabelEncoder()
cols = ['Attrition', 'Travel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
df[cols] = df[cols].apply(le.fit_transform)
#df.drop(['ID', 'EmployeeCount', 'StandardHours', 'Over18'], axis=1, inplace=True)

X = df.loc[:, df.columns != "Attrition"]
y = df['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_model = SVC(kernel = 'poly', class_weight = 'balanced')
svm_model.fit(X_train_scaled, y_train)

# feature_importances = np.abs(svm_model.coef_[0])
# feature_names = X.columns
# sorted_indices = np.argsort(feature_importances)[::-1]

# N = 15 
# top_features = feature_names[sorted_indices][:N]
# print("Top Features:", top_features)
# X_top = X[top_features]

# X_train, X_test, y_train, y_test = train_test_split(X_top, y, test_size=0.2, random_state=42)

# svm_model_top = SVC(kernel='linear', random_state=42)
# svm_model_top.fit(X_train, y_train)

y_pred = svm_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
df_test = pd.read_csv("gdg_test.csv")
test_cols = ['Travel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
df_test[test_cols] = df_test[test_cols].apply(le.fit_transform)
X_Testing = df_test.loc[:, df_test.columns != 'Attrition']
X_Testing_scaled = scaler.transform(X_Testing)
# X_Testing = X_Testing[top_features]
predictions = svm_model.predict(X_Testing_scaled)
df_test['Predicted_label'] = le.inverse_transform(predictions)
df_test[['ID', 'Predicted_label']].to_csv('gdg_test_predictions7.csv', index = False)
df_test
```
